```python
import matplotlib.pyplot as plt
import numpy as np

# Configuraci√≥n inicial
a√±os = list(range(2023, 2044))
gastos_anuales = []

# Simulaci√≥n de gastos
for a√±o in a√±os:
    gastos_mensuales = np.random.uniform(100, 140, 12)  # Generamos 12 valores aleatorios para cada mes
    gasto_anual = gastos_mensuales.sum()
    gastos_anuales.append(gasto_anual)

# Calculo de gastos acumulados
gastos_acumulados = np.cumsum(gastos_anuales)

# Gr√°fico de l√≠nea
plt.figure(figsize=(10,6))
plt.plot(a√±os, gastos_acumulados, marker='o')
plt.title('Gasto acumulado por familia (2023-2043)')
plt.xlabel('A√±o')
plt.ylabel('Gasto Acumulado (‚Ç¨)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Configuraci√≥n inicial
a√±os = list(range(2023, 2044))
unidades_familiares = 1179
gastos_anuales_totales = []

# Simulaci√≥n de gastos
for a√±o in a√±os:
    gastos_mensuales = np.random.uniform(100, 140, 12)  # Generamos 12 valores aleatorios para cada mes
    gasto_anual = gastos_mensuales.sum() * unidades_familiares  # Multiplicamos por las unidades familiares
    gastos_anuales_totales.append(gasto_anual)

# Gr√°fico de l√≠nea
plt.figure(figsize=(10,6))
plt.plot(a√±os, gastos_anuales_totales, marker='o')
plt.title('Gasto anual total por todas las familias (2023-2043)')
plt.xlabel('A√±o')
plt.ylabel('Gasto Anual Total (‚Ç¨)')
plt.ylim(0, 2000000)  # Ajuste de la escala desde 0 hasta 2 millones
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Calculo de gastos acumulados
gastos_acumulados_totales = np.cumsum(gastos_anuales_totales)

# Gr√°fico de l√≠nea
plt.figure(figsize=(10,6))
plt.plot(a√±os, gastos_acumulados_totales, marker='o')
plt.title('Gasto acumulado por todas las familias (2023-2043)')
plt.xlabel('A√±o')
plt.ylabel('Gasto Acumulado Total (‚Ç¨)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Datos iniciales
potencia_pico = 250  # MWp
horas_solares_pico = 2250  # horas al a√±o
tasa_degradacion = 0.005  # 0.5% al a√±o

# Lista para almacenar la energ√≠a generada cada a√±o
energias = []

# Calcular la energ√≠a generada cada a√±o teniendo en cuenta la degradaci√≥n
energia_inicial = potencia_pico * horas_solares_pico
for a√±o in range(2026, 2047):
    energias.append(energia_inicial)
    energia_inicial *= (1 - tasa_degradacion)  # Disminuir la energ√≠a generada por la tasa de degradaci√≥n

# Graficar
a√±os = list(range(2026, 2047))
plt.figure(figsize=(10, 6))
plt.plot(a√±os, energias, marker='o')
plt.title('Energ√≠a generada por una central solar de 250 MWp (2026-2046)')
plt.xlabel('A√±o')
plt.ylabel('Energ√≠a generada (MWh)')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Datos proporcionados
a√±os = list(range(2008, 2022))
deuda = [1.1, 1.6, 1.4, 1.1, 2.3, 2.5, 2.1, 2.0, 1.8, 1.5, 1.1, 1.9, 1.6, 1.8]

# Extendiendo datos hasta que la deuda sea 0
a√±o_actual = 2022
while deuda[-1] > 0:
    a√±os.append(a√±o_actual)
    if a√±o_actual <= 2024:
        deuda.append(deuda[-1])  # Mantener la deuda igual hasta 2024
    else:
        deuda.append(max(0, deuda[-1] - 0.3))  # Reducir la deuda en 0.3 millones cada a√±o a partir de 2025
    a√±o_actual += 1

# Creando el gr√°fico
plt.figure(figsize=(12, 7))
plt.plot(a√±os, deuda, marker='o', color='b', linestyle='-')
plt.title('Deuda P√∫blica en Millones de Euros')
plt.xlabel('A√±o')
plt.ylabel('Deuda en Millones de Euros')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()

# Mostrando el gr√°fico
plt.show()
```
